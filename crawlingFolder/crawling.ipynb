```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


# 넷플릭스 영화 순위 페이지의 URL
url = "https://pedia.watcha.com/ko-KR"

# 페이지에 접속하여 HTML 코드를 가져옴
response = requests.get(url)
html = response.text

# BeautifulSoup을 이용하여 HTML 코드를 파싱
soup = BeautifulSoup(html, 'html.parser')

parent_tag = soup.find_all(class_='css-1qq59e8')
netflix_tag = parent_tag[2]

results = []
base_url = 'https://pedia.watcha.com'
for idx, movie_li in enumerate(netflix_tag.find_all(class_='css-8y23cj')):
    movie_link = movie_li.find('a')['href']
    # css_10hm9rg = movie_li.find(class_='css-10hm9rg')
    movie_title = movie_li.find(class_='css-5yuqaa').text
    # movie_year_country = movie_li.find(class_='css-1rxwuxd').text
    movie_mean_rate = movie_li.find(class_='average css-xgmur2').find_all('span')[1].text

    movie = {
        'rank' : idx + 1,
        'movies_link': base_url + movie_link,
        'title': movie_title,
        'mean_rate': movie_mean_rate
    }

    movie_url = movie["movies_link"]
    # 페이지에 접속하여 HTML 코드를 가져옴
    movie_response = requests.get(movie_url)
    movie_html = movie_response.text
    # BeautifulSoup을 이용하여 HTML 코드를 파싱
    movie_soup = BeautifulSoup(movie_html, 'html.parser')

    parent_tag = movie_soup.find_all('article')[0]
    # parent_tag
    title = parent_tag.find(class_='css-wvh1uf-Summary eokm2781').contents[0].strip()
    year_country_genre = parent_tag.find_all(class_='css-1t00yeb-OverviewMeta eokm2782')[0].text.split('·')
    year = year_country_genre[0].strip()
    country = year_country_genre[1].strip()
    genre = year_country_genre[2].strip()
    try:
        time_regulation = parent_tag.find_all(class_='css-1t00yeb-OverviewMeta eokm2782')[1].text.split('·')
        time = time_regulation[0].strip()
        regulation = time_regulation[1].strip()
    except IndexError:
        time = "Unknown"
        regulation = "Unknown"
        
    description = parent_tag.find(class_='css-kywn6v-StyledText eb5y16b1').text.strip()

    movie["year"] = year
    movie["country"] = country
    movie["genre"] = genre
    movie["time"] = time
    movie["regulation"] = regulation
    movie["description"] = description
    
    results.append(movie)


# results 딕셔너리 객체를 DataFrame으로 변환
df = pd.DataFrame(results)

# DataFrame을 CSV 파일로 저장
df.to_csv("movie_rankings.csv", index=False, encoding='utf-8-sig')
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


class MovieScraper:
    def __init__(self, base_url):
        self.base_url = base_url
        self.movies = []

    def get_movies(self, url):
        response = requests.get(url)
        html = response.text
        soup = BeautifulSoup(html, 'html.parser')

        parent_tag = soup.find_all(class_='css-1qq59e8')
        netflix_tag = parent_tag[2]

        for idx, movie_li in enumerate(netflix_tag.find_all(class_='css-8y23cj')):
            movie_link = movie_li.find('a')['href']
            movie_title = movie_li.find(class_='css-5yuqaa').text
            movie_mean_rate = movie_li.find(class_='average css-xgmur2').find_all('span')[1].text

            movie = {
                'rank': idx + 1,
                'movies_link': self.base_url + movie_link,
                'title': movie_title,
                'mean_rate': movie_mean_rate
            }

            self.get_movie_details(movie)
            self.movies.append(movie)

    def get_movie_details(self, movie):
        movie_url = movie["movies_link"]
        response = requests.get(movie_url)
        html = response.text
        soup = BeautifulSoup(html, 'html.parser')

        parent_tag = soup.find_all('article')[0]
        title = parent_tag.find(class_='css-wvh1uf-Summary eokm2781').contents[0].strip()
        year_country_genre = parent_tag.find_all(class_='css-1t00yeb-OverviewMeta eokm2782')[0].text.split('·')
        year = year_country_genre[0].strip()
        country = year_country_genre[1].strip()
        genre = year_country_genre[2].strip()
        
        try:
            time_regulation = parent_tag.find_all(class_='css-1t00yeb-OverviewMeta eokm2782')[1].text.split('·')
            time = time_regulation[0].strip()
            regulation = time_regulation[1].strip()
        except IndexError:
            time = "Unknown"
            regulation = "Unknown"

        description = parent_tag.find(class_='css-kywn6v-StyledText eb5y16b1').text.strip()

        movie["year"] = year
        movie["country"] = country
        movie["genre"] = genre
        movie["time"] = time
        movie["regulation"] = regulation
        movie["description"] = description

    def save_to_csv(self, file_name):
        df = pd.DataFrame(self.movies)
        df.to_csv(file_name, index=False, encoding='utf-8-sig')


if __name__ == "__main__":
    base_url = "https://pedia.watcha.com"
    ranking_url = "https://pedia.watcha.com/ko-KR"
    scraper = MovieScraper(base_url)
    scraper.get_movies(ranking_url)
    scraper.save_to_csv("movie_rankings.csv")
```

```python
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import pandas as pd


class ReviewScraper:
    def __init__(self, movie_url, executable_path="./chromedriver"):
        self.movie_url = movie_url
        self.executable_path = executable_path
        self.reviews = []

    def get_reviews(self):
        movie_comment_link = self.movie_url + '/comments'
        driver = webdriver.Chrome(executable_path=self.executable_path)
        driver.get(movie_comment_link)

        prev_height = driver.execute_script("return document.body.scrollHeight")

        while len(self.reviews) < 200:
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")
            time.sleep(1)

            current_height = driver.execute_script("return document.body.scrollHeight")

            if current_height == prev_height:
                break

            prev_height = current_height
            page_source = driver.page_source
            soup = BeautifulSoup(page_source, "html.parser")
            ul_tag = soup.find(class_="css-10n5vg9-VisualUl ep5cwgq0")
            bawlbm_tags = ul_tag.find_all(class_="css-bawlbm")

            for review_object in bawlbm_tags:
                reviewer_rate = review_object.find(class_="css-4obf01")
                review_text = review_object.find(class_="css-4tkoly")

                if reviewer_rate and review_text:
                    try:
                        comment = {}
                        comment['reviewer'] = reviewer_rate.find('a')['title']
                        comment['rate'] = reviewer_rate.find(class_='css-yqs4xl').text
                        comment['review'] = review_text.text.replace('\n', '')

                        if comment not in self.reviews:  # Prevent duplicates
                            self.reviews.append(comment)
                    except AttributeError:
                        continue

                if len(self.reviews) >= 200:
                    break

        driver.quit()

    def save_to_csv(self, movie_title, file_name):
        df = pd.DataFrame(self.reviews)
        df['movie_title'] = movie_title  # Add movie_title as a new column
        df.to_csv(file_name, index=False, encoding='utf-8-sig')


if __name__ == "__main__":
    movie_info = pd.read_csv("movie_rankings.csv")
    
    all_reviews = []

    for idx, row in movie_info.iterrows():
        movie_title = row['title']
        movie_url = row['movies_link']
        scraper = ReviewScraper(movie_url)
        scraper.get_reviews()

        for review in scraper.reviews:
            review['movie_title'] = movie_title

        all_reviews.extend(scraper.reviews)

    reviews_df = pd.DataFrame(all_reviews)
    reviews_df.to_csv("reviews_total.csv", index=False, encoding='utf-8-sig')
```
